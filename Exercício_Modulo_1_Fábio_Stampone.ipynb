## 4\. Booleanos

Utilize a tabela da verdade para responder: qual o valor da variável x?

```
a = False
b = True

x = not a & b
```

Na tabela verdade do And (&), o valor True só é válido quando o valor das duas variáveis também forem TRUE.

A variável a, originalmente, recebe o valor FALSE

Já a variável b, recebe o valor TRUE

Ao negar o valor da variável A, temos que A = True, visto que houve uma inversão.

Logo: x = True & True = True.

```
a = False
b = True

x = not a & b
print(x)
```
