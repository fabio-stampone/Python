#### 1) A base 'sinasc_RO_2019.csv' contém dados do sistema de nascidos vivos do DataSUS, cujo dicionário de dados está no arquivo 'Estrutura_SINASC_para_CD.pdf'. Carregue esses arquivos.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib widget

df = pd.read_csv("SINASC_RO_2019.csv")

df.fillna(9, inplace=True)

df.head()
```

####  2) Monte um gráficos de barras com as variáveis APGAR1, APGAR5, SEXO, GESTACAO, ESTCIVMAE.

```python
variables = ['APGAR1', 'APGAR5', 'SEXO', 'GESTACAO', 'ESTCIVMAE']

# Criar um gráfico de barras para cada variável
for var in variables:
    plt.figure(figsize=(10, 6))
    df[var].value_counts().plot(kind='bar', color='skyblue')
    plt.title(f'Distribuição de {var}')
    plt.xlabel(var)
    plt.ylabel('Contagem')
    plt.xticks(rotation=45)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
# Calcular o APGAR5 médio por CONSULTAS
apgar5_mean_by_consultas = df.groupby('CONSULTAS')['APGAR5'].mean()

# Plotar o gráfico de barras
plt.figure(figsize=(10, 6))
apgar5_mean_by_consultas.plot(kind='bar', color='skyblue')
plt.title('APGAR5 Médio por CONSULTAS')
plt.xlabel('CONSULTAS')
plt.ylabel('APGAR5 Médio')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Gráfico 1: Frequência de SEXO
df['SEXO'].value_counts().plot(kind='bar', ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title('Frequência de SEXO')
axes[0, 0].set_xlabel('SEXO')
axes[0, 0].set_ylabel('Contagem')
axes[0, 0].grid(axis='y', linestyle='--', alpha=0.7)

# Gráfico 2: Frequência de GESTACAO
df['GESTACAO'].value_counts().plot(kind='bar', ax=axes[0, 1], color='lightgreen')
axes[0, 1].set_title('Frequência de GESTACAO')
axes[0, 1].set_xlabel('GESTACAO')
axes[0, 1].set_ylabel('Contagem')
axes[0, 1].grid(axis='y', linestyle='--', alpha=0.7)

# Gráfico 3: Frequência de ESTCIVMAE
df['ESTCIVMAE'].value_counts().plot(kind='bar', ax=axes[1, 0], color='lightcoral')
axes[1, 0].set_title('Frequência de ESTCIVMAE')
axes[1, 0].set_xlabel('ESTCIVMAE')
axes[1, 0].set_ylabel('Contagem')
axes[1, 0].grid(axis='y', linestyle='--', alpha=0.7)

# Gráfico 4: Frequência de CONSULTAS
df['CONSULTAS'].value_counts().plot(kind='bar', ax=axes[1, 1], color='gold')
axes[1, 1].set_title('Frequência de CONSULTAS')
axes[1, 1].set_xlabel('CONSULTAS')
axes[1, 1].set_ylabel('Contagem')
axes[1, 1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
